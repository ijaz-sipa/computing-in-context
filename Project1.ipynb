Data Vizualization

```python
import csv

retail_price = []
scaled_prices = []

with open("Fruit-Prices-2022.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        price = float(row["RetailPrice"])
        retail_price.append(price)

# Standardizing price
min_price = min(retail_price)
max_price = max(retail_price)

# constructed the codes below using Chatgpt
for price in retail_price:
    scaled_price = (price - min_price) / (max_price - min_price) * 10
    scaled_prices.append(scaled_price) 

              
sparkline = []
with open("Fruit-Prices-2022.csv", mode='r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row, price in zip(reader, scaled_prices):
        fruit = row["Fruit"]
        sparkline.append(f"{fruit}: {'🍎' * int(price)}")

# printing vizuals
for line in sparkline:
    print(line)
```
